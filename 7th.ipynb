```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Scores of ten students in SMIP and DBMS
x = [70, 46, 54, 34, 20, 86, 18, 12, 56, 64, 42]
y = [60, 66, 90, 46, 16, 48, 74, 8, 32, 54, 62]

# Create a DataFrame
df = pd.DataFrame({'SMIP': x, 'DBMS': y})

# Compute the Spearman rank correlation
corr, pval = stats.spearmanr(df['SMIP'], df['DBMS'])

# Print correlation coefficient and p-value
print("Correlation coefficient:", corr)
print("p-value:", pval)

# Plot heatmap
corr_matrix = df.corr(method='spearman')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation Heatmap')
plt.show()

# Plot histogram of the Spearman rank correlation matrix
plt.hist(corr_matrix.values.flatten(), bins=5, edgecolor='k', alpha=0.7)
plt.title('Histogram of Spearman Rank Correlation Coefficients')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Frequency')
plt.show()
```
