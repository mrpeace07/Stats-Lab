```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv("AusAntidiabeticDrug.csv")  # Replace with the actual file name

# Convert 'ds' column to datetime
df['ds'] = pd.to_datetime(df['ds'])

# Set the style for seaborn
sns.set(style="whitegrid", color_codes=True)

# Plot the data
plt.figure(figsize=(16, 6))
plt.plot(df['ds'], df['y'])
plt.xlabel("Time")
plt.ylabel("$ Millions")
plt.title("Antibiotic Drug Sales Over Time")
plt.show()
```
